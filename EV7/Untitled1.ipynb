```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from PIL import Image
import os
import math
from google.colab import files

modelo = MobileNetV2(weights='imagenet')

# Función para preprocesar la imagen que subas
def preprocesar_imagen(imagen):
    imagen = imagen.resize((224, 224))
    imagen_array = img_to_array(imagen)
    imagen_array = np.expand_dims(imagen_array, axis=0)
    imagen_array = preprocess_input(imagen_array)
    return imagen_array

# Subir una imagen
uploaded = files.upload()

# Procesar las imágenes subidas
for nombre_imagen in uploaded.keys():
    imagen = Image.open(nombre_imagen)
    imagen_preprocesada = preprocesar_imagen(imagen)

    # Hacer la predicción
    prediccion = modelo.predict(imagen_preprocesada)

    # Decodificar la predicción
    resultado = decode_predictions(prediccion, top=3)[0]

    # Mostrar resultados
    print(f'Predicciones para la imagen {nombre_imagen}:')
    for i, (id_clase, etiqueta, confianza) in enumerate(resultado):
        print(f'  {i+1}: {etiqueta} con {confianza:.4f} confianza')

    # Mostrar la imagen
    plt.imshow(imagen)
    plt.axis('off')
    plt.show()
```
